```python
import matplotlib.pyplot as plt
import decimal
import numpy as np

xmin = -10
xmax = 10
dx = 0.1

xlist = np.around(np.arange(xmin, xmax, dx))

ylist = 1 / xlist
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xlist, ylist)
plt.show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-3, 3, 3)
Y = np.arange(-3, 3, 3)
X, Y = np.meshgrid(X, Y)
Z =  X + Y
Z2 = X + Y + 10
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```

```python

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 100, 1)
Y = np.arange(-5, 100, 1)
X, Y = np.meshgrid(X, Y)
Z =  X**3 + Y**3
Z2 = X**(2) + Y**4 + 10
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```
